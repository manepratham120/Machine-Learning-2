```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from imblearn.over_sampling import SMOTE
```

```python
df=pd.read_csv("/content/creditcard.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df.shape
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
df.Class.value_counts()
```

```python
legal= df[df.Class==0]
fraud= df[df.Class==1]
```

```python
legal.shape
```

```python
fraud.shape
```

```python
legal.describe()
```

```python
fraud.describe()
```

```python
fraud.shape
```

```python
legal_sample=legal.sample(n=108)
```

```python
legal_sample.shape
```

```python
data_undersample=pd.concat([legal_sample,fraud],ignore_index=True)
```

```python
data_undersample
```

```python
data_undersample.Class.value_counts()
```

```python
x=data_undersample.drop(columns=['Class'])
y=data_undersample["Class"]
```

```python
x.shape
```

```python
y.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
scalar=StandardScaler()
```

```python
x_train_scalar=scalar.fit_transform(X_train)
x_test_scalar1=scalar.fit_transform(X_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train_scalar,Y_train)
```

```python
pred=model.predict(x_test_scalar1)
```

```python
pred
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test,pred)
```

```python
# oversampling

sm=SMOTE(random_state=42)
```

```python
x_oversampling , y_oversampling=sm.fit_resample(x,y.ravel())
```

```python
pd.value_counts(y_oversampling)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x_oversampling,y_oversampling,test_size=0.2,random_state=42)
```

```python
x_train_scalar=scalar.fit_transform(X_train)
x_test_scalar1=scalar.fit_transform(X_test)
```

```python
model=LogisticRegression()
```

```python
model.fit(x_train_scalar,Y_train)
```

```python
predict=model.predict(x_test_scalar1)
```

```python
predict
```

```python
accuracy_score(Y_test,predict)
```
